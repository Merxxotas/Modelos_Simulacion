## Familia <u>No Congruencial</u>

### <u>Cuadrado Medio</u>

![Cuadrado Medio](https://i.ibb.co/tm7YMWb/Cuadrado-Medio.png)

```python
def obtener_entrada_valida(mensaje, tipo):
    while True:
        try:
            valor = tipo(input(mensaje))
            if tipo == int and valor <= 0:
                print("Por favor, ingrese un número positivo.")
                continue
            return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def resaltar_seleccion(numero_str, inicio, longitud):
    return f"{numero_str[:inicio]}[{numero_str[inicio:inicio+longitud]}]{numero_str[inicio+longitud:]}"

def cuadrado_medio(semilla, d, iteraciones):
    resultados = []
    x = semilla

    for _ in range(iteraciones):
        # Paso 1: Calcular Yi+1 = Xi^2
        y = x * x

        # Convertir y a string para manipular los dígitos
        y_str = str(y)
        y_str_original = y_str  # Guardar el valor original para mostrarlo

        # Determinar si necesitamos agregar ceros al inicio
        if len(y_str) < d * 2:
            y_str = y_str.zfill(d * 2)
            y_str_mostrar = f"0{y_str_original}"  # Para mostrar explícitamente el cero añadido
        else:
            y_str_mostrar = y_str_original

        # Paso 2: Obtener Xi+1
        inicio = 0
        if len(y_str) > d * 2:
            inicio = (len(y_str) - d) // 2
        elif len(y_str) == d * 2:
            inicio = 1

        x_nuevo = int(y_str[inicio:inicio + d])

        # Resaltar la selección de dígitos
        y_str_resaltado = resaltar_seleccion(y_str_mostrar, inicio, d)

        # Calcular Ri
        r = x_nuevo / (10 ** d)

        # Guardar resultados
        resultados.append({
            'Xi': x,
            'Yi+1': y,
            'Yi+1_str': y_str_resaltado,
            'Xi+1': x_nuevo,
            'Ri': r
        })

        x = x_nuevo

    return resultados

def imprimir_resultados(resultados):
    for i, r in enumerate(resultados, 1):
        print(f"\nIteración {i}:")
        print(f"Xi = {r['Xi']}")
        print(f"Yi+1 = {r['Yi+1']} → {r['Yi+1_str']}")
        print(f"Xi+1 = {r['Xi+1']}")
        print(f"Ri = {r['Ri']:.4f}")

def main():
    print("Generador de números pseudoaleatorios - Método del cuadrado medio")
    print("-------------------------------------------------------------")

    # Obtener entradas del usuario
    semilla = obtener_entrada_valida("Ingrese el número semilla (Xo): ", int)
    d = obtener_entrada_valida("Ingrese el número de dígitos (D): ", int)
    iteraciones = obtener_entrada_valida("Ingrese el número de iteraciones: ", int)

    print(f"\nIniciando con:")
    print(f"Semilla (Xo) = {semilla}")
    print(f"Dígitos (D) = {d}")

    resultados = cuadrado_medio(semilla, d, iteraciones)
    imprimir_resultados(resultados)

if __name__ == "__main__":
    main()
```

### <u>Producto Medio</u>

![Producto Medio](https://i.ibb.co/dkwNMYC/Producto-Medio.png)

```python
def obtener_entrada_valida(mensaje, tipo):
    while True:
        try:
            valor = tipo(input(mensaje))
            if tipo == int and valor <= 0:
                print("Por favor, ingrese un número positivo.")
                continue
            return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def resaltar_seleccion(numero_str, inicio, longitud):
    return f"{numero_str[:inicio]}[{numero_str[inicio:inicio+longitud]}]{numero_str[inicio+longitud:]}"

def producto_medio(x0, x1, d, iteraciones):
    resultados = []
    xi = x0
    xi_1 = x1

    for _ in range(iteraciones):
        # Paso 1: Calcular Yi+2 = Xi * Xi+1
        y = xi * xi_1

        # Convertir y a string para manipular los dígitos
        y_str = str(y)
        y_str_original = y_str

        # Determinar si necesitamos agregar ceros al inicio
        if len(y_str) < d + 4:  # Necesitamos al menos D+4 dígitos (2 en cada extremo + D en el medio)
            y_str = y_str.zfill(d + 4)
            y_str_mostrar = f"0{y_str_original}"
        else:
            y_str_mostrar = y_str_original

        # Paso 2: Obtener Xi+2
        # Siempre dejamos 2 dígitos en cada extremo
        inicio = 2
        if len(y_str) > d + 4:
            # Si hay más dígitos de los necesarios, ajustamos el inicio
            inicio = (len(y_str) - d - 4) // 2 + 2

        xi_2 = int(y_str[inicio:inicio + d])

        # Resaltar la selección de dígitos
        y_str_resaltado = resaltar_seleccion(y_str_mostrar, inicio, d)

        # Calcular Ri
        r = xi_2 / (10 ** d)

        # Guardar resultados
        resultados.append({
            'Xi': xi,
            'Xi+1': xi_1,
            'Yi+2': y,
            'Yi+2_str': y_str_resaltado,
            'Xi+2': xi_2,
            'Ri': r
        })

        # Preparar para la siguiente iteración
        xi = xi_1
        xi_1 = xi_2

    return resultados

def imprimir_resultados(resultados):
    for i, r in enumerate(resultados, 1):
        print(f"\nIteración {i}:")
        if i == 1:
            print(f"X0 = {r['Xi']}")
            print(f"X1 = {r['Xi+1']}")
        else:
            print(f"Xi = {r['Xi']}")
            print(f"Xi+1 = {r['Xi+1']}")
        print(f"Yi+2 = {r['Yi+2']} → {r['Yi+2_str']}")
        print(f"Xi+2 = {r['Xi+2']}")
        print(f"Ri = {r['Ri']:.4f}")

def main():
    print("Generador de números pseudoaleatorios - Método del Producto medio")
    print("-------------------------------------------------------------")

    # Obtener entradas del usuario
    x0 = obtener_entrada_valida("Ingrese el primer número semilla (X0): ", int)
    x1 = obtener_entrada_valida("Ingrese el segundo número semilla (X1): ", int)
    d = obtener_entrada_valida("Ingrese el número de dígitos (D): ", int)
    iteraciones = obtener_entrada_valida("Ingrese el número de iteraciones: ", int)

    print(f"\nIniciando con:")
    print(f"X0 = {x0}")
    print(f"X1 = {x1}")
    print(f"Dígitos (D) = {d}")

    resultados = producto_medio(x0, x1, d, iteraciones)
    imprimir_resultados(resultados)

if __name__ == "__main__":
    main()
```

### <u>Multiplicador Constante</u>


![Multiplicador Constante](https://i.ibb.co/jHGcxS0/Multiplicador-Constante.png)

```python
def obtener_entrada_valida(mensaje, tipo):
    while True:
        try:
            valor = tipo(input(mensaje))
            if tipo == int and valor <= 0:
                print("Por favor, ingrese un número positivo.")
                continue
            return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def resaltar_seleccion(numero_str, inicio, longitud):
    return f"{numero_str[:inicio]}[{numero_str[inicio:inicio+longitud]}]{numero_str[inicio+longitud:]}"

def multiplicador_constante(a, x0, d, iteraciones):
    resultados = []
    xi = x0

    for i in range(iteraciones):
        # Paso 1: Calcular Yi+1 = a * Xi
        y = a * xi

        # Convertir y a string para manipular los dígitos
        y_str = str(y)
        y_str_original = y_str

        # Determinar si necesitamos agregar ceros al inicio
        if len(y_str) < d + 4:  # Necesitamos al menos D+4 dígitos (2 en cada extremo + D en el medio)
            y_str = y_str.zfill(d + 4)
            y_str_mostrar = f"0{y_str_original}"
        else:
            y_str_mostrar = y_str_original

        # Paso 2: Obtener Xi+1
        # Siempre dejamos 2 dígitos en cada extremo
        inicio = 2
        if len(y_str) > d + 4:
            # Si hay más dígitos de los necesarios, ajustamos el inicio
            inicio = (len(y_str) - d - 4) // 2 + 2

        xi_next = int(y_str[inicio:inicio + d])

        # Resaltar la selección de dígitos
        y_str_resaltado = resaltar_seleccion(y_str_mostrar, inicio, d)

        # Calcular Ri
        r = xi_next / (10 ** d)

        # Guardar resultados
        resultados.append({
            'i': i,
            'Xi': xi,
            'Yi+1': y,
            'Yi+1_str': y_str_resaltado,
            'Xi+1': xi_next,
            'Ri': r
        })

        # Preparar para la siguiente iteración
        xi = xi_next

    return resultados

def imprimir_resultados(resultados, a):
    print(f"\nMultiplicador constante a = {a}")
    for r in resultados:
        i = r['i']
        print(f"\nIteración {i+1}:")
        print(f"X{i} = {r['Xi']}")
        print(f"Y{i+1} = aX{i} = {a} × {r['Xi']} = {r['Yi+1']} → {r['Yi+1_str']}")
        print(f"X{i+1} = {r['Xi+1']}")
        print(f"R{i+1} = {r['Ri']:.4f}")

def main():
    print("Generador de números pseudoaleatorios - Método del Multiplicador Constante")
    print("-------------------------------------------------------------")

    # Obtener entradas del usuario
    a = obtener_entrada_valida("Ingrese el multiplicador constante (a): ", int)
    x0 = obtener_entrada_valida("Ingrese el número semilla (X0): ", int)
    d = obtener_entrada_valida("Ingrese el número de dígitos (D): ", int)
    iteraciones = obtener_entrada_valida("Ingrese el número de iteraciones: ", int)

    print(f"\nIniciando con:")
    print(f"Multiplicador constante (a) = {a}")
    print(f"X0 = {x0}")
    print(f"Dígitos (D) = {d}")

    resultados = multiplicador_constante(a, x0, d, iteraciones)
    imprimir_resultados(resultados, a)

if __name__ == "__main__":
    main()
```

### <u>Lineal</u>


![Lineal](https://i.ibb.co/7N5R4bL/Lineal.png)

```python
def obtener_entrada_valida(mensaje, tipo, condiciones=None):
    while True:
        try:
            valor = tipo(input(mensaje))
            if condiciones:
                for condicion, mensaje_error in condiciones:
                    if not condicion(valor):
                        print(mensaje_error)
                        break
                else:  # Si todas las condiciones se cumplen
                    return valor
            else:
                return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def verificar_a(k):
    return 1 + 4 * k

def congruencial_lineal(x0, k, c, g):
    # Calcular parámetros
    m = 2 ** g
    a = verificar_a(k)

    # Verificar que los parámetros cumplan las condiciones
    if x0 <= 0:
        raise ValueError("X0 debe ser un entero positivo")
    if k <= 0:
        raise ValueError("K debe ser un entero positivo")
    if c <= 0 or c % 2 == 0:
        raise ValueError("C debe ser un entero positivo impar")
    if g <= 0:
        raise ValueError("g debe ser un entero positivo")

    return {
        'm': m,
        'a': a,
        'k': k,
        'c': c,
        'g': g,
        'x0': x0
    }

def generar_secuencia(params, iteraciones):
    resultados = []
    xi = params['x0']
    m = params['m']
    a = params['a']
    c = params['c']

    for i in range(iteraciones):
        # Calcular Xi+1
        xi_next = (a * xi + c) % m
        # Calcular Ri+1
        ri_next = xi_next / (m - 1)

        # Guardar resultados
        resultados.append({
            'i': i,
            'Xi': xi,
            'Xi+1': xi_next,
            'Ri+1': ri_next,
            'calculo': f"({a} × {xi} + {c}) mod {m} = {a * xi + c} mod {m} = {xi_next}"
        })

        xi = xi_next

    return resultados

def imprimir_resultados(params, resultados):
    print("\nParámetros iniciales:")
    print(f"m = 2^{params['g']} = {params['m']}")
    print(f"a = 1 + 4k = 1 + 4({params['k']}) = {params['a']}")
    print(f"k = {params['k']}")
    print(f"c = {params['c']}")
    print(f"g = {params['g']}")
    print(f"X0 = {params['x0']}")

    print("\nSecuencia generada:")
    for r in resultados:
        i = r['i']
        print(f"\nIteración {i+1}:")
        print(f"X{i+1} = (aX{i} + C) mod m = {r['calculo']}")
        print(f"R{i+1} = X{i+1}/(m-1) = {r['Xi+1']}/{params['m']-1} = {r['Ri+1']:.3f}")

def main():
    print("Generador de números pseudoaleatorios - Método Congruencial Lineal")
    print("-------------------------------------------------------------")

    # Obtener parámetros del usuario
    x0 = obtener_entrada_valida("Ingrese la semilla inicial (X0): ", int,
                               [(lambda x: x > 0, "X0 debe ser un entero positivo")])
    k = obtener_entrada_valida("Ingrese el valor de k: ", int,
                              [(lambda x: x > 0, "k debe ser un entero positivo")])
    c = obtener_entrada_valida("Ingrese el valor de c (debe ser impar): ", int,
                              [(lambda x: x > 0, "c debe ser un entero positivo"),
                               (lambda x: x % 2 == 1, "c debe ser impar")])
    g = obtener_entrada_valida("Ingrese el valor de g: ", int,
                              [(lambda x: x > 0, "g debe ser un entero positivo")])

    iteraciones = obtener_entrada_valida("Ingrese el número de iteraciones: ", int,
                                        [(lambda x: x > 0, "El número de iteraciones debe ser positivo")])

    try:
        params = congruencial_lineal(x0, k, c, g)
        resultados = generar_secuencia(params, iteraciones)
        imprimir_resultados(params, resultados)
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

### <u>Lineal</u>


![Multiplicativo](https://i.ibb.co/B6bFF4w/Multiplicativo.png)

```python
def obtener_entrada_valida(mensaje, tipo, condiciones=None):
    while True:
        try:
            valor = tipo(input(mensaje))
            if condiciones:
                for condicion, mensaje_error in condiciones:
                    if not condicion(valor):
                        print(mensaje_error)
                        break
                else:  # Si todas las condiciones se cumplen
                    return valor
            else:
                return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def verificar_a(k, tipo):
    if tipo == 3:
        return 8 * k + 3
    else:  # tipo == 5
        return 8 * k + 5

def congruencial_multiplicativo(x0, k, g, tipo_a):
    # Calcular parámetros
    m = 2 ** g
    a = verificar_a(k, tipo_a)

    # Verificar que los parámetros cumplan las condiciones
    if x0 <= 0 or x0 % 2 == 0:
        raise ValueError("X0 debe ser un entero positivo impar")
    if k < 0:
        raise ValueError("K debe ser un entero no negativo")
    if g <= 2:
        raise ValueError("g debe ser un entero mayor a 2")

    # Calcular el número garantizado de valores distintos
    valores_garantizados = m // 4

    return {
        'm': m,
        'a': a,
        'k': k,
        'g': g,
        'x0': x0,
        'valores_garantizados': valores_garantizados,
        'tipo_a': tipo_a
    }

def generar_secuencia(params):
    resultados = []
    xi = params['x0']
    m = params['m']
    a = params['a']
    valores_garantizados = params['valores_garantizados']

    valores_generados = set()
    i = 0

    while True:
        # Calcular Xi+1
        xi_next = (a * xi) % m
        ri_next = xi_next / (m - 1)

        # Guardar resultados
        resultados.append({
            'i': i,
            'Xi': xi,
            'Xi+1': xi_next,
            'Ri+1': ri_next,
            'calculo': f"({a} × {xi}) mod {m} = {a * xi} mod {m} = {xi_next}"
        })

        valores_generados.add(xi_next)

        # Verificar si hemos completado un ciclo
        if xi_next in {resultado['Xi'] for resultado in resultados[:-1]}:
            break

        xi = xi_next
        i += 1

    return resultados, len(valores_generados)

def imprimir_resultados(params, resultados, valores_unicos):
    print("\nParámetros iniciales:")
    print(f"g = {params['g']}")
    print(f"k = {params['k']}")
    print(f"m = 2^g = 2^{params['g']} = {params['m']}")
    print(f"a = 8k + {params['tipo_a']} = 8 × {params['k']} + {params['tipo_a']} = {params['a']}")
    print(f"X0 = {params['x0']}")
    print(f"\nSe garantizan m/4 = 2^{params['g']}/4 = {params['m']}/4 = {params['valores_garantizados']} números pseudoaleatorios distintos")

    print("\nSecuencia generada:")
    for r in resultados:
        i = r['i']
        print(f"\nIteración {i+1}:")
        print(f"X{i+1} = {r['calculo']}")
        print(f"R{i+1} = X{i+1}/(m-1) = {r['Xi+1']}/{params['m']-1} = {r['Ri+1']:.4f}")

    print(f"\nSe generaron {valores_unicos} valores únicos antes de que el ciclo se repitiera")
    if valores_unicos >= params['valores_garantizados']:
        print("✓ Se cumple la garantía de valores distintos")
    else:
        print("⚠ No se cumple la garantía de valores distintos")

def main():
    print("Generador de números pseudoaleatorios - Método Congruencial Multiplicativo")
    print("-------------------------------------------------------------")

    # Obtener parámetros del usuario
    x0 = obtener_entrada_valida("Ingrese la semilla inicial (X0, debe ser impar): ", int,
                               [(lambda x: x > 0, "X0 debe ser un entero positivo"),
                                (lambda x: x % 2 == 1, "X0 debe ser impar")])
    k = obtener_entrada_valida("Ingrese el valor de k: ", int,
                              [(lambda x: x >= 0, "k debe ser un entero no negativo")])
    g = obtener_entrada_valida("Ingrese el valor de g (debe ser mayor a 2): ", int,
                              [(lambda x: x > 2, "g debe ser un entero mayor a 2")])

    tipo_a = obtener_entrada_valida("Seleccione el tipo de 'a' (3 para 8k+3, 5 para 8k+5): ", int,
                                   [(lambda x: x in [3, 5], "Debe seleccionar 3 o 5")])

    try:
        params = congruencial_multiplicativo(x0, k, g, tipo_a)
        resultados, valores_unicos = generar_secuencia(params)
        imprimir_resultados(params, resultados, valores_unicos)
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

### <u>Aditivo</u>


![Aditivo](https://i.ibb.co/PZrsnYy/Aditivo.png)

```python
def obtener_entrada_valida(mensaje, tipo, condiciones=None):
    while True:
        try:
            valor = tipo(input(mensaje))
            if condiciones:
                for condicion, mensaje_error in condiciones:
                    if not condicion(valor):
                        print(mensaje_error)
                        break
                else:  # Si todas las condiciones se cumplen
                    return valor
            else:
                return valor
        except ValueError:
            print(f"Por favor, ingrese un {'número entero' if tipo == int else 'valor'} válido.")

def obtener_semillas(k):
    semillas = []
    for i in range(k):
        while True:
            semilla = obtener_entrada_valida(f"Ingrese la semilla X{i+1}: ", int,
                                            [(lambda x: x > 0, "La semilla debe ser un entero positivo")])
            if semilla in semillas:
                print("Esta semilla ya ha sido ingresada. Las semillas deben ser distintas.")
            else:
                semillas.append(semilla)
                break
    return semillas

def congruencial_aditivo(semillas, k, m, iteraciones):
    if len(semillas) != k:
        raise ValueError(f"Se requieren exactamente {k} semillas iniciales")
    if len(set(semillas)) != k:
        raise ValueError("Todas las semillas deben ser distintas")
    if any(x <= 0 for x in semillas):
        raise ValueError("Todas las semillas deben ser positivas")
    if m <= 0:
        raise ValueError("m debe ser un entero positivo")

    resultados = []
    valores = semillas.copy()

    for i in range(iteraciones):
        # Calcular Xi+1
        nuevo_valor = (valores[i] + valores[i-k+1]) % m
        ri = nuevo_valor / (m-1) if m > 1 else 0

        # Guardar resultados
        resultados.append({
            'i': i+k+1,
            'Xi': valores[i],
            'Xi-k+1': valores[i-k+1],
            'Xi+1': nuevo_valor,
            'Ri': ri,
            'calculo': f"({valores[i]} + {valores[i-k+1]}) mod {m} = {valores[i] + valores[i-k+1]} mod {m} = {nuevo_valor}"
        })

        valores.append(nuevo_valor)

    return resultados, valores

def imprimir_resultados(semillas, k, m, resultados, valores):
    print("\nParámetros iniciales:")
    for i, semilla in enumerate(semillas, 1):
        print(f"X{i} = {semilla}")
    print(f"k = {k}")
    print(f"m = {m}")

    print("\nSecuencia generada:")
    for i, semilla in enumerate(semillas, 1):
        print(f"X{i} = {semilla} (semilla inicial)")

    for r in resultados:
        i = r['i']
        print(f"\nX{i} = {r['calculo']}")
        print(f"R{i-k} = X{i}/(m-1) = {r['Xi+1']}/{m-1} = {r['Ri']:.4f}")

    # Detectar ciclo
    secuencia = valores[k:]
    for i in range(len(secuencia)):
        for j in range(i+1, len(secuencia)):
            if secuencia[i:i+k] == secuencia[j:j+k]:
                print(f"\nSe detectó un ciclo de longitud {j-i}")
                return

def main():
    print("Generador de números pseudoaleatorios - Método Congruencial Aditivo")
    print("-------------------------------------------------------------")

    k = obtener_entrada_valida("Ingrese el valor de k (número de semillas): ", int,
                              [(lambda x: x > 0, "k debe ser un entero positivo")])

    semillas = obtener_semillas(k)

    m = obtener_entrada_valida("Ingrese el valor de m: ", int,
                              [(lambda x: x > 0, "m debe ser un entero positivo")])

    iteraciones = obtener_entrada_valida("Ingrese el número de iteraciones: ", int,
                                        [(lambda x: x > 0, "El número de iteraciones debe ser positivo")])

    try:
        resultados, valores = congruencial_aditivo(semillas, k, m, iteraciones)
        imprimir_resultados(semillas, k, m, resultados, valores)
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
